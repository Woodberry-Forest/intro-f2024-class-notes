````python
with open("./odyssey.txt", "r") as f:
    text = f.read()
print(text)
type(text)
````

````python
letters = {}

for letter in text:
    if letter not in letters.keys():
        letters[letter] = 1
    else:
        letters[letter] += 1

print(letters)
````

````python
letters_list = list(letters.items())
````

````python
letters_list
````

````python
for key, value in sorted(letters_list, key=lambda x: x[1], reverse=True):
    print(key, value)
````

````python
len(text)
````

````python
lets = []
counts = []
for key, value in sorted(letters_list, key=lambda x: x[1], reverse=True):
    lets.append(key)
    counts.append(value)
````

````python
from matplotlib import pyplot as plt
````

````python
plt.bar(lets, counts)
````
